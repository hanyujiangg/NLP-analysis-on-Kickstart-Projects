```python
import pandas as pd 
our_data = pd.read_excel('KSdataset(Cleaned).xlsx',sheet_name = 'Sheet1',header =0,encoding="utf8")
```

```python
# pd.set_option('display.max_colwidth', -1)
import numpy as np
our_data['Comment_URL'] = our_data['Comment_URL'].str.strip(' https://www.kickstarter.com/projects/')
```

```python
our_data = our_data.drop('Image', 1)
```

```python
our_data['project'] = our_data.Comment_URL.map(lambda x:x.split('/')[1])
```

```python
len(our_data.project.unique())
```

```python
our_data['project_name'] = our_data.project.str.replace('-',' ')
our_data
```

```python
project_name = our_data['project_name'].unique()
project_name[0]
```

```python
project_slug = our_data['project'].unique()
project_slug[0]
```

```python
import os 
all_other_files = []
root = 'other-source-dataset'
folders = os.listdir('other-source-dataset')
for folder in folders:
    if folder != '.DS_Store':
        folder_path = os.path.join(root, folder)
        for file in os.listdir(folder_path):
            all_other_files.append(os.path.join(folder_path,file))
```

```python
all_other_files
```

```python
# other_data = pd.read_csv('other-source-dataset/')Ã·
other_df = pd.concat(map(pd.read_csv, all_other_files))
```

```python
other_df['name'].nunique()
```

```python
other_df['name'].duplicated().any()
```

```python
import ast 
other_df['profile_dict'] = other_df['profile'].map(lambda x:ast.literal_eval(x))
```

```python
other_df['project_id'] = other_df['profile'].apply(lambda x:x['project_id'])
```

```python
other_df['processed_name'] = other_df['name'].str.lower()
```

```python
other_df
```

```python
count = 0 
for name in project_name:
    print(name)
    if name in other_df['processed_name']:
        count += 1
print(count)
```

```python
other_df_selected = other_df[['backers_count','blurb','category','converted_pledged_amount','country','country_displayable_name','currency','current_currency','fx_rate','goal','id','location','name','pledged','profile','slug','state','static_usd_rate','urls','usd_pledged']]



```

```python
result_df = other_df_selected.drop_duplicates(subset=['id'], keep='first')
```

```python
result_df
```

```python
result_df['has'] = result_df['slug'].apply(lambda x: x in project_slug)
```

```python
result_df.drop_duplicates(subset=['id'], keep='first')
```

```python
result_df.reset_index(drop=True)
```

```python
result_df2 = result_df[result_df['has'] == True]
```

```python
result_df2.id
```

```python
our_data['project_id'] = our_data['project'].apply(lambda x: result_df2['id'])
```

```python
final_df = our_data.merge(result_df2,how='inner',left_on = 'project', right_on='slug')
```

```python
final_df.to_csv('final_data.csv', index=False)
```
